# Comparison with de Wilde & de Bie, 2000 data

```python
import numpy as np
import pandas as pd
import datetime
import matplotlib.pyplot as plt
```

```python
rdata = 'D:/GitHub/PlanktonOfSchelde/data/deWildedeBie/'
```

### Importing 1978

```python
OCT78 = pd.read_csv(rdata+'schelde-FN2O-OCT78.csv') #µmol/m2/day
```

```python
OCT78
```

### Importing 1993

```python
OCT93 = pd.read_csv(rdata+'schelde-FN2O-OCT93.csv') #µmol/m2/day
```

### Importing 1996

```python
JUL96 = pd.read_csv(rdata+'schelde-FN2O-JUL96.csv') #µmol/m2/day
```

### Importing compartment area

```python
pdAREA = pd.read_csv('D:/GitHub/PlanktonOfSchelde/data/schelde_compartment_deWilde.csv')
```

```python
pdAREA #m2
```

```python
totOCT78 = 0
totOCT93 = 0
totJUL96 = 0
for i in range(len(OCT78.x)) :
    if i >= 2 :
        totOCT78 += OCT78[' y'][i]*pdAREA.area[round(OCT78.x[i])-1]
        totOCT93 += OCT93[' y'][i]*pdAREA.area[round(OCT93.x[i])-1]
        totJUL96 += JUL96[' y'][i]*pdAREA.area[round(JUL96.x[i])-1]
```

### Belgica data

```python
belgica = pd.read_csv(rdata+'Belgica2023.csv')
```

```python
belgica
```

```python
sumBelgica = np.sum(belgica.area*belgica['flux_stations_Flux µmol.m**-2.hour**-1']*24)
```

```python
print('Flux en mars 2023 %.2f kmol.m$^{-2}.day$^{-1}$' %(sumBelgica/10**9))
```

```python
plt.scatter(datetime.datetime(1978,10,1),totOCT78/(10**9), c='b', marker='+',s=50)
plt.text(datetime.datetime(1978,10,1),totOCT78/(10**9)-1.5, 'F$_{N_2O}$ = %.2f kmol.day$^{-1}$' %(totOCT78/(10**9)))

plt.scatter(datetime.datetime(1993,10,1),totOCT93/(10**9), c='b', marker='+',s=50)
plt.text(datetime.datetime(1993,10,1),totOCT93/(10**9)-1.5, 'F$_{N_2O}$ = %.2f kmol.day$^{-1}$' %(totOCT93/(10**9)))


plt.scatter(datetime.datetime(1996,7,1),totJUL96/(10**9), c='b', marker='+',s=50)
plt.text(datetime.datetime(1996,7,1),totJUL96/(10**9)-1.5, 'F$_{N_2O}$ = %.2f kmol.day$^{-1}$' %(totJUL96/(10**9)))

plt.scatter(datetime.datetime(2023,3,10),sumBelgica/(10**9), c='r', marker='^')
plt.text(datetime.datetime(2007,5,10),sumBelgica/(10**9)+1.0, 'F$_{N_2O}$ = %.2f kmol.day$^{-1}$' %(sumBelgica/(10**9)),c='r')

plt.xlabel('Year')
plt.ylabel('Total flux of N$_2$O kmol.day$^{-1}$')
```
